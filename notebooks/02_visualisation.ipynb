Then I needed to explore obtained data. I started with visualizing all the time series. All time series contained missing values and records were not taken with constant time step. Different variables happened to have different time points when the data were recorded. I visualized the data in two ways:/
1. using points, with no line where data were missing
2. joining all known values with lines

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../data/df_all.csv', parse_dates=['date'], date_format='%Y-%m-%d %H:%M:%S')
df.dtypes
```

```python
df.head(2)
```

```python
df.set_index('date', inplace=True)
```

```python
df.plot(title='all')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()
```

```python
for col in df.columns:
    df[col].dropna().plot()
plt.title('all')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
```
