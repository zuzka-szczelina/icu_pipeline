## Correlation matrix

I tried to briefly explore correlation between parameters. For this purpose I created a scaterrplot matrix. Slight correlation was visible between systolic and diastolic pressure (which was not surprising). 

```python
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.plotting import scatterplotmatrix
```

```python
df = pd.read_csv('../../data/df.csv', parse_dates=['date'], date_format='%Y-%m-%d %H:%M:%S')
df_lin_int = pd.read_csv('../../data/df_linearly_interpolated.csv', parse_dates=['date'], date_format='%Y-%m-%d %H:%M:%S')

df_lin_int.set_index('date', inplace=True)
```

```python
df.dtypes
```

```python
scatterplotmatrix(df_lin_int.values, figsize=(12, 10), names=df_lin_int.columns, alpha=0.5, s=5)
plt.tight_layout()
plt.show() 
```
