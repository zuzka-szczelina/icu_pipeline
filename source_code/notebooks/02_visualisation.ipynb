## Visualisation 

Then I needed to explore obtained data. I started with visualizing all the time series. All time series contained missing values and records were not taken with constant time step. Different parameters happened to have different time points when the data were recorded. I visualized the data in two ways:/
1. using points, with no line connection at places where some data were missing
2. joining all known values with straight lines

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../../data/df_all.csv', parse_dates=['date'], date_format='%Y-%m-%d %H:%M:%S')
df.dtypes
```

```python
df.head(2)
```

```python
df.set_index('date', inplace=True)
```

```python
df.plot(title='all')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()
```

```python
for col in df.columns:
    df[col].dropna().plot()
plt.title('all')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
```

```python
df.columns
```

```python
df[['breaths_all', 'breaths_forced']].plot(title='breaths')
plt.show()
```

```python
df[['RR_systolic', 'RR_diastolic']].plot(title='arterial blood pressure', marker='.')
plt.show()
df_plot = df[['RR_systolic', 'RR_diastolic']].dropna() # .dropna(axis='index', how='any') - same, default
df_plot.plot(title='arterial blood pressure')
plt.show()
```

```python
df[['pulse']].plot(title='pulse')
plt.show()
df_plot = df[['pulse']].dropna()
df_plot.plot(title='pulse')
plt.show()
```
