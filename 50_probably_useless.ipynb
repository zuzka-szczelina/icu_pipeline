```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/df.csv',parse_dates=['date'], date_format='%Y-%m-%d %H:%M:%S')
df.set_index('date', inplace=True)
df.head(3)
```

I have a dataframe with time recors taken on full hour or "some hour" + 30 mins. First I fill dataframe to have rows every 30 mins

```python
df_30mins = df.resample('30min').asfreq()
```

Than, I linearly interpolate all time series:

```python
df_interpolated = df_30mins.interpolate(method='linear')
df_interpolated.head(10)
```

than I choose only full hour records:

```python
df_hourly = df_interpolated.resample('h').asfreq()
df_hourly
```
